## Perform the data cleaning operations mentioned so far in class

### Delete the column education and the number of open complaints from the dataframe.

```python
datanew=data.drop(['Education','Number of Open Complaints'],axis=1)
```

```python
datanew.head()
```

```python
datanew.info()
```

### Correct the values in the column customer lifetime value. They are given as a percent, so multiply them by 100 and change dtype to numerical type.

```python
datanew['Customer Lifetime Value']
```

```python
datanew['Customer Lifetime Value']=pd.to_numeric(data['Customer Lifetime Value'],errors='coerce')
```

```python
# Customer lifetime value changed to numeric value
datanew.info()
```

```python
#Multiplied Customer lifetime value x 100 
datanew["Customer Lifetime Value"]=100 * datanew["Customer Lifetime Value"]
```

```python
datanew["Customer Lifetime Value"]
```

### Check for duplicate rows in the data and remove if any.
